```python
#import libraries

import pandas as pd #pip install pandas
import numpy as np #pip install numpy
import pickle
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier

df = pd.read_csv('Titanic Dataset.csv')
df[100:110]

```

```python
df.describe
```

```python
df.columns
```

```python
len(df)
```

```python
df.drop(columns=['PassengerId','Name','Ticket'],inplace=True)
```

```python
df.head(5)
```

```python
df['Survived'].unique()
```

```python
d = {"Pclass":"Tier", "Sex":"Gender", "SibSp":"Number of Siblings/Spouses Aboard","Parch":"Number of Parents/Children Aboard"}
df.rename(columns=d, inplace=True)
```

```python
df.head(5)
```

```python
df.isna().sum()
```

```python
df.dropna(subset=["Age"])
```

```python
df['Age'].fillna(df['Age'].median(),inplace=True)
df['Cabin'].fillna('General Cabin',inplace=True)
df['Fare'].fillna(df['Fare'].median(),inplace=True)
```

```python
df.head(5)
```

```python
df.isna().sum()
```

```python
#Encoding part
le = LabelEncoder()
df['Gender'] = le.fit_transform(df['Gender']) #male 0, female 1
df['Cabin'] = le.fit_transform(df['Cabin']) #general cabin 76
df['Embarked'] = le.fit_transform(df['Embarked']) #Q -1 , S-2, C-0
df.head()
```

```python
ss = StandardScaler()  #10000, age - 30-45, 90%, 10% - outliers 90, 10, 5, 85, standardize the numbers 

df[['Age','Fare']] = ss.fit_transform(df[['Age','Fare']])
df.head(10)
```

```python
#dataset splitting into two parts, training and testing
X = df[['Tier','Gender','Age','Number of Siblings/Spouses Aboard','Number of Parents/Children Aboard','Fare','Cabin','Embarked']]#feature
y = df['Survived']#label or target variable
```

```python
#creating train and test dataset 100%, test_size, keep aside 20% as testing data
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2)
```

```python
model = LogisticRegression()
model.fit(X_train,y_train)
```

```python
#test the model
y_pred = model.predict(X_test)
print(y_pred)
```

```python
#importing accuracy models
from sklearn.metrics import accuracy_score, confusion_matrix
```

```python
#accuracy metrics - >
print(accuracy_score(y_test,y_pred))
```

```python
print(confusion_matrix(y_test,y_pred))
```

```python
model_dt = DecisionTreeClassifier()
model_dt.fit(X_train,y_train)
```

```python
model_rf = RandomForestClassifier()
model_rf.fit(X_train,y_train)
```
